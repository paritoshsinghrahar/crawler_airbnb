```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

class InstaBot:
    def __init__(self, username, pw):
        self.driver = webdriver.Chrome(executable_path = 'H:\Documents\Downloads\chromedriver')
        self.username = username
        self.driver.get('https://www.instagram.com/')
        self.driver.implicitly_wait(2)
        self.driver.find_element_by_name("username").send_keys(username)
        self.driver.find_element_by_name("password").send_keys(pw)
        self.driver.find_element_by_xpath('//button[@type="submit"]').click()
        self.driver.implicitly_wait(4)

    def not_now(self):
        self.driver.find_element_by_xpath("//button[contains(text(), 'Not Now')]").click()
        self.driver.implicitly_wait(2)
        
    def search_food(self):
        self.driver.find_element_by_xpath('//input[@type="text"]').send_keys('food')
        ans = self.driver.find_elements_by_xpath('//span[@class="Ap253"]')
        for i in ans:
            text_ig = i.text
            if text_ig[0] != '#':
                print(text_ig)
        
    def open_page_sodelhi(self):
        self.driver.find_element_by_xpath('//input[@type="text"]').clear()
        self.driver.find_element_by_xpath('//input[@type="text"]').send_keys('So Delhi')
        self.driver.implicitly_wait(8)
        self.driver.find_element_by_xpath("//a[contains(@href,'/sodelhi')]").click()
    
    def open_page_dilsefoodie(self):
        self.driver.find_element_by_xpath('//input[@type="text"]').clear()
        self.driver.find_element_by_xpath('//input[@type="text"]').send_keys('dilsefoodie')
        self.driver.implicitly_wait(8)
        self.driver.find_element_by_xpath("//a[contains(@href,'/dilsefoodie')]").click()
    
    def open_page_foodtalkindia(self):
        self.driver.find_element_by_xpath('//input[@type="text"]').clear()
        self.driver.find_element_by_xpath('//input[@type="text"]').send_keys('foodtalkindia')
        self.driver.implicitly_wait(8)
        self.driver.find_element_by_xpath("//a[contains(@href,'/foodtalkindia')]").click()
        
    def follow_page(self):
        try:
            if self.driver.find_element_by_xpath("//span[@aria-label='Following']").is_displayed():
                print('Already Followed')
        except:
            self.driver.find_element_by_xpath("//button[contains(text(), 'Follow')]").click()
            print('Followed')


    def unfollow_page(self):
        self.driver.refresh()

        try:
            if self.driver.find_element_by_xpath("//button[contains(text(), 'Follow')]").is_displayed():
                print('Already Unfollowed')
        except:
            self.driver.find_element_by_xpath("//span[@aria-label='Following']").click()
            self.driver.find_element_by_xpath("//button[contains(text(), 'Unfollow')]").click()
            print('Unfollowed')
    
    def like_unlike30(self):
        time.sleep(5)
        ans = self.driver.find_elements_by_xpath('//div[@class="_9AhH0"]')
        for i in range(30):
            try:
                ans[i].click()
                wait = WebDriverWait(self.driver, 50)
                wait.until(EC.presence_of_element_located((By.XPATH, 
                                                      '//div[@class="o-MQd  z8cbW"]')))
                btn = self.driver.find_elements_by_xpath('//button[@class="wpO6b "]')
                btn[2].click()
                btn[-1].click()
            except:
                ans = self.driver.find_elements_by_xpath('//div[@class="_9AhH0"]')
                ans[i].click()
                wait = WebDriverWait(self.driver, 50)
                wait.until(EC.presence_of_element_located((By.XPATH, 
                                                      '//div[@class="o-MQd  z8cbW"]')))
                btn = self.driver.find_elements_by_xpath('//button[@class="wpO6b "]')
                btn[2].click()
                btn[-1].click()

    def following(self,driver,flag):
        followingLink = self.driver.find_elements_by_css_selector('ul li a')
        followingLink[1].click()
        followingList = self.driver.find_element_by_css_selector('div[role=\'dialog\'] ul')
        numberOffollowingInList = len(followingList.find_elements_by_css_selector('li'))

        from selenium.webdriver.common.keys import Keys
        import time

        followingList.click()
        actionChain = webdriver.ActionChains(self.driver)
        time.sleep(3)

        while (numberOffollowingInList < 24):
            time.sleep(2)
            actionChain.key_down(Keys.SPACE).perform()
            followingList.click()
            actionChain.key_down(Keys.SPACE).perform()
            actionChain.key_down(Keys.SPACE).perform()
            actionChain.key_down(Keys.SPACE).perform()

            prevnumberOffollowingInList = numberOffollowingInList
            numberOffollowingInList = len(followingList.find_elements_by_css_selector('li'))
            
            if numberOffollowingInList == prevnumberOffollowingInList:
                break

        while (numberOffollowingInList < 200):
            time.sleep(2)
            actionChain.key_down(Keys.SPACE).perform()
            actionChain.key_down(Keys.SPACE).perform()
            actionChain.key_down(Keys.SPACE).perform()
            actionChain.key_down(Keys.SPACE).perform()
            
            prevnumberOffollowingInList = numberOffollowingInList
            numberOffollowingInList = len(followingList.find_elements_by_css_selector('li'))
            if numberOffollowingInList == prevnumberOffollowingInList:
                break
        

        for user in followingList.find_elements_by_css_selector('li'):
            userLink = user.find_element_by_css_selector('a').get_attribute('href')
            if userLink[26:-1] == 'inspireaphor':
                flag=0
                break
            
        return flag
        
    def followers_500(self,following_check=0):
        
        followersLink = self.driver.find_element_by_css_selector('ul li a')
        followersLink.click()
        followersList = self.driver.find_element_by_css_selector('div[role=\'dialog\'] ul')
        numberOfFollowersInList = len(followersList.find_elements_by_css_selector('li'))

        followersList.click()
        actionChain = webdriver.ActionChains(self.driver)
        time.sleep(3)

        while (numberOfFollowersInList < 24):
            time.sleep(2)
            actionChain.key_down(Keys.SPACE).perform()
            followersList.click()
            actionChain.key_down(Keys.SPACE).perform()
            actionChain.key_down(Keys.SPACE).perform()
            actionChain.key_down(Keys.SPACE).perform()

            numberOfFollowersInList = len(followersList.find_elements_by_css_selector('li'))
        
        while (numberOfFollowersInList < 500):
            time.sleep(2)
            actionChain.key_down(Keys.SPACE).perform()
            actionChain.key_down(Keys.SPACE).perform()
            actionChain.key_down(Keys.SPACE).perform()
            actionChain.key_down(Keys.SPACE).perform()

            numberOfFollowersInList = len(followersList.find_elements_by_css_selector('li'))

        followers = []
        for user in followersList.find_elements_by_css_selector('li'):
            userLink = user.find_element_by_css_selector('a').get_attribute('href')
            print(userLink[26:-1])
            followers.append(userLink)
            
            buttonx = user.find_elements_by_xpath('//div[@class="Pkbci"]/button')
            ans_username= str(buttonx[2].get_attribute('innerHTML'))
            
            if str == 'Following':
                print(ans_username)
                print(userLink[26:-1])

            if (len(followers) == 500):
                break
                
        print('No. of followers:', len(followers))
        buttony = self.driver.find_elements_by_xpath('//button[@class="sqdOP  L3NKy    _8A5w5    "]')
        print('The followers you follow are:',print(len(buttony)-1))

        if following_check == 1:
            buttonm = self.driver.find_elements_by_xpath('//button[@class="sqdOP  L3NKy    _8A5w5    "]/parent::*/parent::*')
            if len(buttonm) > 1:

                link=[]
                usernames=[]
                for i in range(1,len(buttonm)):
                    #print(buttonm[i].find_element_by_xpath('div/div[2]/div/span/a').get_attribute('outerHTML'))
                    link.append(buttonm[i].find_element_by_xpath('div/div[2]/div/span/a').get_attribute('href'))
                    usernames.append(buttonm[i].find_element_by_xpath('div/div[2]/div/span/a').get_attribute('text'))

                for i in range(1,len(buttonm)):
                    self.driver.get(link[i-1])
                    flag = following(self.driver,1)
                    if flag == 0 :
                        print('Mutual Friends Both Follow each other')
                        print(usernames[i-1])
                    else:
                        print("Don't follow back")
                        print(usernames[i-1])
            else:
                print('You follow no one in top 500')

            
    def story_coding(self):
        self.driver.get('https://www.instagram.com/coding.ninjas/')
        time.sleep(10)
        try:
            self.driver.find_element_by_xpath("//span[@class='_2dbep ']").click()
            print("Story Open")
        except:
            print("No Story Available")
            
    def refreshit(self):
        self.driver.refresh()
```
